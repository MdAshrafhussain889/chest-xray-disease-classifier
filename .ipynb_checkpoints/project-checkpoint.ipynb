````python
# Cell 4: Load CSV and parse disease labels
print("="*80)
print("CELL 4: LOAD AND PARSE DATASET")
print("="*80)

# Load CSV file
print("\n✓ Loading Data Entry CSV...")
data_df = pd.read_csv(DATA_ENTRY_FILE)
print(f"  - Total records: {len(data_df)}")
print(f"  - Columns: {list(data_df.columns)}")

# Display first few rows
print(f"\n  First 3 rows:")
print(data_df.head(3))

# Parse findings column
print("\n✓ Parsing disease findings...")

def parse_findings(finding_str):
    """Convert pipe-separated findings string to binary vector"""
    findings = str(finding_str).split('|')
    labels = np.zeros(NUM_CLASSES)
    for finding in findings:
        finding = finding.strip()
        # Handle case variations for Pleural_Thickening
        if finding in ["Pleural_Thickening", "Pleural_thickening"]:
            finding = "Pleural_Thickening"
        if finding in DISEASE_CLASSES:
            idx = DISEASE_CLASSES.index(finding)
            labels[idx] = 1
    return labels

# Apply parsing
print("  - Encoding labels...")
data_df['encoded_labels'] = data_df['Finding Labels'].apply(parse_findings)

# Convert to arrays
image_paths = data_df['Image Index'].values
labels = np.array(data_df['encoded_labels'].tolist())

print(f"\n✓ Data loaded and encoded!")
print(f"  - Total images: {len(image_paths)}")
print(f"  - Label shape: {labels.shape}")

# Display class distribution
print("\n✓ Class Distribution:")
class_counts = labels.sum(axis=0)

# Create DataFrame for better visualization
class_dist_df = pd.DataFrame({
    'Disease': DISEASE_CLASSES,
    'Count': class_counts.astype(int),
    'Percentage': (class_counts / len(image_paths) * 100)
})

print(class_dist_df.to_string(index=False))

# Compute class weights for handling imbalance
print("\n✓ Computing class weights for imbalance handling...")
class_weights = np.zeros(NUM_CLASSES)
for i in range(NUM_CLASSES):
    pos_count = np.sum(labels[:, i])
    if pos_count > 0:
        class_weights[i] = len(labels) / (2.0 * pos_count)
    else:
        class_weights[i] = 1.0

print("\n  Class weights (higher = more weight for rare classes):")
weights_df = pd.DataFrame({
    'Disease': DISEASE_CLASSES,
    'Weight': class_weights
})
print(weights_df.to_string(index=False))

print("\n✓ Data Loading Complete!")
````

````python
# Cell 5: Split data at patient level to prevent data leakage
print("="*80)
print("CELL 5: PATIENT-LEVEL DATA SPLITTING")
print("="*80)

# Extract patient ID from image filename (take the first token before underscore)
print("\n✓ Extracting patient IDs...")
# If image_paths contains full paths, use os.path.basename to get filename
import os
patient_ids = np.array([os.path.basename(p).split('_')[0] for p in image_paths])
unique_patients = np.unique(patient_ids)

print(f"  - Total unique patients: {len(unique_patients)}")
print(f"  - Total images: {len(image_paths)}")
print(f"  - Average images per patient: {len(image_paths) / len(unique_patients):.2f}")

# Patient-level split (prevents data leakage)
print("\n✓ Performing patient-level split...")
print("  Strategy: 70% train, 10% validation, 20% test")

# First split: 80% train+val, 20% test
train_val_patients, test_patients = train_test_split(
    unique_patients,
    test_size=0.2,
    random_state=SEED
)

# Second split: 70/10 ratio for train and val from train+val set
train_patients, val_patients = train_test_split(
    train_val_patients,
    test_size=0.125,  # This gives 10% val of total (0.125 * 0.8 = 0.1)
    random_state=SEED
)

# Create masks for each set (these will now be 1-D boolean arrays)
train_mask = np.isin(patient_ids, train_patients)
val_mask   = np.isin(patient_ids, val_patients)
test_mask  = np.isin(patient_ids, test_patients)

# Create image path lists
train_images = image_paths[train_mask]
train_labels = labels[train_mask]

val_images = image_paths[val_mask]
val_labels = labels[val_mask]

test_images = image_paths[test_mask]
test_labels = labels[test_mask]

print(f"\n✓ Split Results:")
print(f"  - Train: {len(train_images):>6} images from {len(train_patients):>5} patients ({len(train_images)/len(image_paths)*100:.1f}%)")
print(f"  - Val:   {len(val_images):>6} images from {len(val_patients):>5} patients ({len(val_images)/len(image_paths)*100:.1f}%)")
print(f"  - Test:  {len(test_images):>6} images from {len(test_patients):>5} patients ({len(test_images)/len(image_paths)*100:.1f}%)")
print(f"  - Total: {len(train_images) + len(val_images) + len(test_images):>6} images")

# Verify no patient overlap
print(f"\n✓ Verifying no patient overlap...")
overlap_train_val = len(np.intersect1d(train_patients, val_patients))
overlap_train_test = len(np.intersect1d(train_patients, test_patients))
overlap_val_test = len(np.intersect1d(val_patients, test_patients))

print(f"  - Train ∩ Val:  {overlap_train_val} patients")
print(f"  - Train ∩ Test: {overlap_train_test} patients")
print(f"  - Val ∩ Test:   {overlap_val_test} patients")

if overlap_train_val == 0 and overlap_train_test == 0 and overlap_val_test == 0:
    print("  ✓ No overlap detected - data split is valid!")
else:
    print("  ❌ WARNING: Patient overlap detected!")

print("\n✓ Data Splitting Complete!")
````

````python
# Cell 6: Create data loading and preprocessing functions
print("="*80)
print("CELL 6: BUILD DATA LOADING FUNCTIONS")
print("="*80)

def load_and_preprocess_image(image_filename, label, base_dir, folders, img_size=IMG_SIZE):
    """Load, preprocess, and return image with label"""
    
    # Find image in folders
    image_path = None
    for folder in folders:
        potential_path = os.path.join(base_dir, folder, image_filename)
        if os.path.exists(potential_path):
            image_path = potential_path
            break
    
    if image_path is None:
        # Return black image if not found
        print(f"Warning: Image not found: {image_filename}")
        img = np.zeros((img_size, img_size, 3), dtype=np.float32)
        return img, label
    
    try:
        # Read image as grayscale
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        if img is None:
            print(f"Warning: Could not read: {image_path}")
            img = np.zeros((img_size, img_size), dtype=np.uint8)
    except Exception as e:
        print(f"Error loading {image_path}: {e}")
        img = np.zeros((img_size, img_size), dtype=np.uint8)
    
    # Resize to target size
    img = cv2.resize(img, (img_size, img_size))
    
    # Normalize to [0, 1]
    img = img.astype(np.float32) / 255.0
    
    # Convert grayscale to RGB (replicate channels for DenseNet)
    img = np.stack([img, img, img], axis=-1)
    
    return img, label


def create_tf_dataset(image_filenames, labels, base_dir, folders, 
                      batch_size=BATCH_SIZE, augment=False, shuffle=True):
    """Create TensorFlow dataset with optional augmentation"""
    
    def load_fn(filename, label):
        """Wrapper for TensorFlow py_function"""
        img, label = tf.py_function(
            lambda fn, lbl: load_and_preprocess_image(
                fn.numpy().decode(), 
                lbl.numpy(), 
                base_dir, 
                folders
            ),
            [filename, label],
            [tf.float32, tf.float32]
        )
        img.set_shape([IMG_SIZE, IMG_SIZE, 3])
        label.set_shape([NUM_CLASSES])
        return img, label
    
    # Create dataset
    dataset = tf.data.Dataset.from_tensor_slices((image_filenames, labels))
    
    if shuffle:
        dataset = dataset.shuffle(buffer_size=min(1000, len(image_filenames)), seed=SEED)
    
    # Map loading function
    dataset = dataset.map(load_fn, num_parallel_calls=tf.data.AUTOTUNE)
    
    # Add augmentation if requested
    if augment:
        def augment_fn(img, label):
            # Random horizontal flip
            img = tf.image.random_flip_left_right(img, seed=SEED)
            
            # Random brightness
            img = tf.image.random_brightness(img, 0.15, seed=SEED)
            
            # Random contrast
            img = tf.image.random_contrast(img, 0.85, 1.15, seed=SEED)
            
            # Clip to valid range
            img = tf.clip_by_value(img, 0.0, 1.0)
            
            return img, label
        
        dataset = dataset.map(augment_fn, num_parallel_calls=tf.data.AUTOTUNE)
    
    # Batch and prefetch
    dataset = dataset.batch(batch_size)
    dataset = dataset.prefetch(tf.data.AUTOTUNE)
    
    return dataset


print("\n✓ Data loading functions created:")
print("  - load_and_preprocess_image()")
print("  - create_tf_dataset()")

print("\n✓ Data Loading Functions Complete!")
````

````python
# Cell 7: Create train, validation, and test datasets
print("="*80)
print("CELL 7: CREATE TENSORFLOW DATASETS")
print("="*80)

print("\n✓ Creating datasets...")
print("  This may take a few minutes...")

# Create train dataset with augmentation
print("\n  1/3 Creating train dataset (with augmentation)...")
train_dataset = create_tf_dataset(
    train_images, 
    train_labels,
    BASE_DIR,
    IMAGE_FOLDERS,
    augment=True, 
    shuffle=True
)
print(f"      ✓ Train dataset created: {len(train_dataset)} batches")

# Create validation dataset (no augmentation, no shuffle)
print("\n  2/3 Creating validation dataset...")
val_dataset = create_tf_dataset(
    val_images, 
    val_labels,
    BASE_DIR,
    IMAGE_FOLDERS,
    augment=False, 
    shuffle=False
)
print(f"      ✓ Val dataset created: {len(val_dataset)} batches")

# Create test dataset (no augmentation, no shuffle)
print("\n  3/3 Creating test dataset...")
test_dataset = create_tf_dataset(
    test_images, 
    test_labels,
    BASE_DIR,
    IMAGE_FOLDERS,
    augment=False, 
    shuffle=False
)
print(f"      ✓ Test dataset created: {len(test_dataset)} batches")

# Test: Load one batch to verify
print("\n✓ Testing dataset pipeline...")
try:
    for images_batch, labels_batch in train_dataset.take(1):
        print(f"  - Batch image shape: {images_batch.shape}")
        print(f"  - Batch label shape: {labels_batch.shape}")
        print(f"  - Image dtype: {images_batch.dtype}")
        print(f"  - Label dtype: {labels_batch.dtype}")
        print(f"  - Image range: [{tf.reduce_min(images_batch).numpy():.3f}, {tf.reduce_max(images_batch).numpy():.3f}]")
        print("  ✓ Sample batch loaded successfully!")
except Exception as e:
    print(f"  ❌ Error loading batch: {e}")

print("\n✓ TensorFlow Datasets Created!")
````

````python
# Cell 8: Build DenseNet121 model architecture
print("="*80)
print("CELL 8: BUILD DENSENET121 MODEL")
print("="*80)

from tensorflow.keras.applications import DenseNet121
from tensorflow.keras import Model, layers, optimizers
from tensorflow.keras.regularizers import l2

print("\n✓ Building DenseNet121 model...")

# Load pre-trained DenseNet121 base model
base_model = DenseNet121(
    include_top=False,
    weights='imagenet',
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

print(f"  - Base model loaded: DenseNet121")
print(f"  - Total base parameters: {base_model.count_params():,}")

# Freeze base model for Phase 1
base_model.trainable = False
print(f"  - Base model frozen for Phase 1 training")

# Build custom classification head
print("\n✓ Building custom classification head...")

inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name='input')

# Pass through base model
x = base_model(inputs, training=False)

# Global average pooling
x = layers.GlobalAveragePooling2D(name='global_avg_pool')(x)

# Batch normalization
x = layers.BatchNormalization(name='bn1')(x)

# Dense layer 1
x = layers.Dense(512, activation='relu', kernel_regularizer=l2(1e-5), name='dense_512')(x)
x = layers.Dropout(0.5, name='dropout_1')(x)

# Dense layer 2
x = layers.Dense(256, activation='relu', kernel_regularizer=l2(1e-5), name='dense_256')(x)
x = layers.Dropout(0.3, name='dropout_2')(x)

# Output layer (sigmoid for multi-label)
outputs = layers.Dense(NUM_CLASSES, activation='sigmoid', name='output')(x)

# Create model
model = Model(inputs=inputs, outputs=outputs, name='DenseNet121_MultiLabel')

print(f"\n✓ Model architecture created!")
print(f"  - Input shape: {model.input_shape}")
print(f"  - Output shape: {model.output_shape}")
print(f"  - Trainable parameters (Phase 1): {model.count_params():,}")

# Display model summary
print("\n✓ Model Summary:")
model.summary()

print("\n✓ Model Building Complete!")
````

````python
# Cell 9: Compile model with weighted binary cross-entropy loss
print("="*80)
print("CELL 9: COMPILE MODEL")
print("="*80)

print("\n✓ Setting up weighted binary cross-entropy loss...")

# Create weighted loss function
def weighted_binary_crossentropy(y_true, y_pred):
    """
    Weighted BCE to handle class imbalance.
    Uses class_weights computed earlier.
    """
    weights = tf.constant([class_weights[i] for i in range(NUM_CLASSES)], dtype=tf.float32)
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    weighted_bce = bce * weights
    return tf.reduce_mean(weighted_bce)

print("  - Loss function: Weighted Binary Cross-Entropy")
print(f"  - Class weights applied: {len(class_weights)} classes")

# Compile model
print("\n✓ Compiling model...")

model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss=weighted_binary_crossentropy,
    metrics=[
        'binary_accuracy',
        tf.keras.metrics.AUC(name='auc'),
        tf.keras.metrics.Recall(name='recall'),
        tf.keras.metrics.Precision(name='precision')
    ]
)

print("  ✓ Model compiled with:")
print("    - Optimizer: Adam (lr=1e-4)")
print("    - Loss: Weighted Binary Cross-Entropy")
print("    - Metrics: Binary Accuracy, AUC, Recall, Precision")

print("\n✓ Model Compilation Complete!")
````

````python
# Cell 10: Setup callbacks for training
print("="*80)
print("CELL 10: SETUP TRAINING CALLBACKS")
print("="*80)

from tensorflow.keras import callbacks

print("\n✓ Configuring callbacks...")

# Model checkpoint for Phase 1
checkpoint_phase1 = callbacks.ModelCheckpoint(
    'phase1_best_model.h5',
    monitor='val_auc',
    mode='max',
    save_best_only=True,
    verbose=1
)

# Early stopping
early_stop = callbacks.EarlyStopping(
    monitor='val_auc',
    mode='max',
    patience=10,
    verbose=1,
    restore_best_weights=True
)

# Reduce learning rate on plateau
reduce_lr = callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    mode='min',
    factor=0.5,
    patience=5,
    min_lr=1e-7,
    verbose=1
)

# Combine callbacks
callbacks_phase1 = [checkpoint_phase1, early_stop, reduce_lr]

print("  ✓ Callbacks configured:")
print("    - ModelCheckpoint (save best model based on val_auc)")
print("    - EarlyStopping (patience=10, restore best weights)")
print("    - ReduceLROnPlateau (factor=0.5, patience=5)")

print("\n✓ Callbacks Setup Complete!")
````

````python
# Cell 11: TRAIN PHASE 1 (FROZEN BASE) - Robust, full working cell
print("="*80)
print("CELL 11: TRAINING PHASE 1 (FROZEN BASE)")
print("="*80)

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# ---------------------------
# Helper: weighted loss maker
# ---------------------------
def get_weighted_binary_crossentropy(pos_weight):
    """
    Returns a loss function applying pos_weight to positive class.
    Works with model outputs that are probabilities (sigmoid).
    Handles shape/dtype safely.
    """
    pos_weight = float(pos_weight)

    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.cast(y_pred, tf.float32)

        # Flatten so shapes match: (batch,)
        y_true = tf.reshape(y_true, [-1])
        y_pred = tf.reshape(y_pred, [-1])

        # Per-sample binary crossentropy (works with probs)
        bce = tf.keras.backend.binary_crossentropy(y_true, y_pred)  # shape (batch,)

        # weights: pos -> pos_weight, neg -> 1.0
        weights = tf.where(tf.equal(y_true, 1.0), pos_weight, 1.0)

        weighted_bce = bce * weights
        return tf.reduce_mean(weighted_bce)

    return loss

# ---------------------------
# Compute recommended pos_weight from train_dataset
# ---------------------------
def compute_pos_weight_from_dataset(ds, max_batches=None):
    """
    Counts positive / negative examples in a tf.data dataset (batched).
    Returns pos_weight = (negatives / positives) clipped to >=1.0.
    If dataset yields sample_weight or is complicated, this may need adapting.
    """
    pos = 0
    neg = 0
    for i, (_, y) in enumerate(ds):
        # allow y to be (batch,) or (batch,1)
        y_np = tf.reshape(tf.cast(y, tf.int64), [-1]).numpy()
        pos += int(np.sum(y_np == 1))
        neg += int(np.sum(y_np == 0))
        if max_batches is not None and i + 1 >= max_batches:
            break
    # avoid division by zero
    pos = max(1, pos)
    neg = max(1, neg)
    pos_weight = neg / pos
    # clamp to reasonable range
    pos_weight = float(np.clip(pos_weight, 1.0, 100.0))
    return pos_weight, pos, neg

print("\n✓ Phase 1 Configuration:")
print(f"  - Base model: FROZEN (assumed)")
try:
    trainable_params = sum([tf.size(w).numpy() for w in model.trainable_weights])
except Exception:
    trainable_params = "unknown (model not defined here)"
print(f"  - Trainable parameters: {trainable_params:,}")
print(f"  - Learning rate: 1e-4")
print(f"  - Batch size: {BATCH_SIZE}")
print(f"  - Epochs: {EPOCHS_PHASE1}")
try:
    print(f"  - Train batches: {len(train_dataset)}")
    print(f"  - Val batches:   {len(val_dataset)}")
except Exception:
    print("  - Train/Val batches: dataset lengths not available (tf.data.dataset)")

# ---------------------------
# Sanity-check shapes for one batch
# ---------------------------
print("\n✓ Sampling one batch to inspect shapes/dtypes...")
for x_sample, y_sample in train_dataset.take(1):
    print("  - x_sample.shape:", getattr(x_sample, "shape", tf.shape(x_sample).numpy()))
    print("  - y_sample.shape:", getattr(y_sample, "shape", tf.shape(y_sample).numpy()), "dtype:", y_sample.dtype)
    break

# ---------------------------
# Compute pos_weight and define loss
# ---------------------------
print("\n✓ Computing pos_weight from train dataset (counts positives/negatives)...")
pos_weight, n_pos, n_neg = compute_pos_weight_from_dataset(train_dataset, max_batches=200)  # limit scans for speed
print(f"  - positives: {n_pos}, negatives: {n_neg}, pos_weight (neg/pos): {pos_weight:.3f}")

loss_fn = get_weighted_binary_crossentropy(pos_weight)

# ---------------------------
# Compile model (recompile to ensure new loss is used)
# ---------------------------
print("\n✓ Compiling model with weighted loss and metrics...")
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
model.compile(
    optimizer=optimizer,
    loss=loss_fn,
    metrics=[tf.keras.metrics.AUC(name='auc'), tf.keras.metrics.BinaryAccuracy(name='binary_accuracy')]
)

# ---------------------------
# Prepare callbacks (ensure ModelCheckpoint saves best by val_auc)
# ---------------------------
checkpoint_path = 'phase1_best_model.h5'
checkpoint_cb = ModelCheckpoint(
    checkpoint_path,
    monitor='val_auc',
    mode='max',
    save_best_only=True,
    save_weights_only=False,
    verbose=1
)

earlystop_cb = EarlyStopping(monitor='val_auc', mode='max', patience=10, restore_best_weights=False, verbose=1)

# combine user callbacks_phase1 if exists; otherwise use ours
combined_callbacks = []
if 'callbacks_phase1' in globals() and callbacks_phase1 is not None:
    # ensure we don't duplicate identical checkpoints; but include user's callbacks
    combined_callbacks.extend(callbacks_phase1 if isinstance(callbacks_phase1, (list, tuple)) else [callbacks_phase1])
# Always add our checkpoint and earlystop (if not already present by name)
combined_callbacks.extend([checkpoint_cb, earlystop_cb])

print("\n✓ Starting Phase 1 Training...")
print("  This may take a long time depending on your GPU/CPU.")
print("  " + "="*76)

# ---------------------------
# Train
# ---------------------------
try:
    history_phase1 = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=EPOCHS_PHASE1,
        callbacks=combined_callbacks,
        verbose=1
    )
except tf.errors.InvalidArgumentError as e:
    print("\n❌ InvalidArgumentError during training (likely shape/broadcast issue in loss).")
    print("Error message:", e.message.split("\n")[0] if hasattr(e, "message") else str(e))
    print("\nSuggested fixes:")
    print(" - Ensure your y labels in train_dataset are 0/1 integers and have shape (batch,) or (batch,1).")
    print(" - The loss included reshaping; if your model outputs logits, use a logits-weighted loss variant.")
    raise

print("  " + "="*76)
print("\n✓ Phase 1 Training Complete!")

# ---------------------------
# Results & save/load
# ---------------------------
# Safely access metrics (history keys may vary if training failed early)
val_auc_history = history_phase1.history.get('val_auc', None)
auc_history = history_phase1.history.get('auc', None)

if val_auc_history is not None and len(val_auc_history) > 0:
    best_epoch = int(np.argmax(val_auc_history)) + 1
    best_val_auc = float(np.max(val_auc_history))
    final_val_auc = float(val_auc_history[-1])
    print(f"\n  Results:")
    print(f"    - Best epoch: {best_epoch}")
    print(f"    - Best val_auc: {best_val_auc:.4f}")
    print(f"    - Final val_auc: {final_val_auc:.4f}")
else:
    print("\n  Results: val_auc not found in history (check metrics names).")
    best_epoch = None

print(f"    - Model saved (best): {checkpoint_path}")

# ---------------------------
# Plot training history (if available)
# ---------------------------
print("\n✓ Plotting training history...")
plt.figure(figsize=(15, 5))

# Loss
plt.subplot(1, 3, 1)
loss_hist = history_phase1.history.get('loss', [])
val_loss_hist = history_phase1.history.get('val_loss', [])
if loss_hist:
    plt.plot(loss_hist, label='Train Loss')
if val_loss_hist:
    plt.plot(val_loss_hist, label='Val Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Phase 1: Loss'); plt.legend(); plt.grid(True)

# AUC
plt.subplot(1, 3, 2)
auc_hist = history_phase1.history.get('auc', [])
val_auc_hist = history_phase1.history.get('val_auc', [])
if auc_hist:
    plt.plot(auc_hist, label='Train AUC')
if val_auc_hist:
    plt.plot(val_auc_hist, label='Val AUC')
plt.xlabel('Epoch'); plt.ylabel('AUC'); plt.title('Phase 1: AUC'); plt.legend(); plt.grid(True)

# Accuracy
plt.subplot(1, 3, 3)
acc_hist = history_phase1.history.get('binary_accuracy', [])
val_acc_hist = history_phase1.history.get('val_binary_accuracy', [])
if acc_hist:
    plt.plot(acc_hist, label='Train Acc')
if val_acc_hist:
    plt.plot(val_acc_hist, label='Val Acc')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.title('Phase 1: Accuracy'); plt.legend(); plt.grid(True)

plt.tight_layout()
plt.savefig('phase1_training_history.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n  ✓ Training plot saved: phase1_training_history.png")

# ---------------------------
# Load best model from checkpoint (safe)
# ---------------------------
print("\n✓ Loading best model from checkpoint...")
try:
    model = tf.keras.models.load_model(
        checkpoint_path,
        custom_objects={'loss': loss_fn, 'weighted_binary_crossentropy': loss_fn}  # supply the loss in case it's needed
    )
    print("  ✓ Loaded best model from checkpoint")
except Exception as e:
    print("  ⚠️ Could not load saved model directly with custom loss (exception shown).")
    print("  You can still continue with current `model` in memory. Exception:", str(e))

print("\n✓ Phase 1 Complete!")
````

````python
# Cell 13: TRAIN PHASE 2 (FINE-TUNING) - Multi-label safe
print("="*80)
print("CELL 13: TRAINING PHASE 2 (FINE-TUNING) - MULTI-LABEL SAFE")
print("="*80)

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import callbacks, optimizers
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# ---------------------------
# Multi-label weighted BCE builder (same shape logic as Phase1)
# ---------------------------
def get_multilabel_weighted_bce(pos_weight_vec):
    pos_weight_vec = tf.convert_to_tensor(np.array(pos_weight_vec, dtype=np.float32))
    pos_weight_vec = tf.reshape(pos_weight_vec, [-1])  # (C,)

    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.cast(y_pred, tf.float32)

        # Ensure shapes: (batch, C)
        C = tf.shape(pos_weight_vec)[0]
        y_true = tf.reshape(y_true, [-1, C])
        y_pred = tf.reshape(y_pred, [-1, C])

        # element-wise BCE -> (batch, C)
        bce = tf.keras.backend.binary_crossentropy(y_true, y_pred)

        # weights: (1, C) broadcast across batch
        pw_broadcast = tf.reshape(pos_weight_vec, (1, -1))
        weights = tf.where(tf.equal(y_true, 1.0), pw_broadcast, tf.ones_like(pw_broadcast))

        weighted_bce = bce * weights
        return tf.reduce_mean(weighted_bce)

    return loss

# ---------------------------
# Helper to compute per-class pos_weights (scans dataset; limits batches for speed)
# ---------------------------
def compute_pos_weights_multilabel(ds, num_classes=None, max_batches=200):
    pos_counts = None
    neg_counts = None
    total_samples = 0
    for i, (_, y) in enumerate(ds):
        y_np = tf.cast(y, tf.int64).numpy()
        # infer shape
        if num_classes is None:
            if y_np.ndim == 1:
                C = 1
                y_np = y_np.reshape(-1, 1)
            else:
                C = y_np.shape[1]
        else:
            C = num_classes
            y_np = y_np.reshape(-1, C)
        if pos_counts is None:
            pos_counts = np.zeros(C, dtype=np.int64)
            neg_counts = np.zeros(C, dtype=np.int64)
        pos_counts += np.sum(y_np == 1, axis=0).astype(np.int64)
        neg_counts += np.sum(y_np == 0, axis=0).astype(np.int64)
        total_samples += y_np.shape[0]
        if max_batches is not None and i + 1 >= max_batches:
            break
    pos_counts = np.maximum(pos_counts, 1)
    neg_counts = np.maximum(neg_counts, 1)
    pos_weights = neg_counts.astype(np.float32) / pos_counts.astype(np.float32)
    pos_weights = np.clip(pos_weights, a_min=1.0, a_max=100.0)
    return pos_weights, pos_counts, neg_counts, total_samples

# ---------------------------
# Infer num_classes from a batch
# ---------------------------
print("\n✓ Inspecting a batch to infer num_classes...")
for xb, yb in train_dataset.take(1):
    y_shape = getattr(yb, "shape", tf.shape(yb).numpy())
    if isinstance(y_shape, (tuple, list)):
        num_classes = int(y_shape[-1]) if len(y_shape) > 1 else 1
    else:
        y_shape_eval = tf.shape(yb).numpy()
        num_classes = int(y_shape_eval[-1]) if y_shape_eval.size > 1 else 1
    print("  - y_sample.shape:", y_shape, "-> num_classes =", num_classes)
    break

# ---------------------------
# Compute pos_weights from train_dataset (reuse Phase1 scanning strategy)
# ---------------------------
print("\n✓ Computing per-class pos_weights (this scans up to 200 batches for speed)...")
pos_weights, pos_counts, neg_counts, scanned = compute_pos_weights_multilabel(train_dataset, num_classes=num_classes, max_batches=200)
for i, (pw, p, n) in enumerate(zip(pos_weights, pos_counts, neg_counts)):
    print(f"    - class {i}: pos={p}, neg={n}, pos_weight={pw:.3f}")
print(f"  - scanned batches (approx): {scanned}")

# ---------------------------
# Build loss and recompile model (lower LR for fine-tuning)
# ---------------------------
loss_fn = get_multilabel_weighted_bce(pos_weights)
print("\n✓ Recompiling model for Phase 2 (lr=1e-5) with multi-label weighted loss...")
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-5),
    loss=loss_fn,
    metrics=[tf.keras.metrics.AUC(name='auc'), tf.keras.metrics.BinaryAccuracy(name='binary_accuracy')]
)

# ---------------------------
# Prepare Phase 2 callbacks (ensure checkpoint_by_val_auc present)
# ---------------------------
checkpoint_p2 = ModelCheckpoint(
    'phase2_best_model.h5',
    monitor='val_auc',
    mode='max',
    save_best_only=True,
    verbose=1
)
earlystop_p2 = EarlyStopping(monitor='val_auc', mode='max', patience=15, restore_best_weights=True, verbose=1)
reduce_lr_p2 = ReduceLROnPlateau(monitor='val_loss', mode='min', factor=0.5, patience=5, min_lr=1e-8, verbose=1)

# combine any existing callbacks_phase2 but ensure our checkpoint is included
combined_callbacks_p2 = []
if 'callbacks_phase2' in globals() and callbacks_phase2 is not None:
    combined_callbacks_p2.extend(callbacks_phase2 if isinstance(callbacks_phase2, (list, tuple)) else [callbacks_phase2])
# Avoid duplicate ModelCheckpoint with same filepath; always ensure ours exists
combined_callbacks_p2.extend([checkpoint_p2, earlystop_p2, reduce_lr_p2])

# ---------------------------
# Train Phase 2
# ---------------------------
print("\n✓ Starting Phase 2 Training...")
print("  This may take a long time depending on hardware.")
print("  " + "="*76)

try:
    history_phase2 = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=EPOCHS_PHASE2,
        callbacks=combined_callbacks_p2,
        initial_epoch=0,
        verbose=1
    )
except tf.errors.InvalidArgumentError as e:
    print("\n❌ InvalidArgumentError during Phase 2 training (likely broadcasting/shape mismatch in loss).")
    print("Error summary:", str(e).split("\n")[0])
    print("Suggestions:")
    print(" - Confirm y in train_dataset has shape (batch, num_classes) and values 0/1.")
    print(" - If your model outputs logits (no sigmoid), replace the loss with a logits-version using tf.nn.weighted_cross_entropy_with_logits.")
    raise

print("  " + "="*76)
print("\n✓ Phase 2 Training Complete!")

# ---------------------------
# Results & plotting
# ---------------------------
val_auc_hist = history_phase2.history.get('val_auc', [])
if val_auc_hist:
    best_epoch_p2 = int(np.argmax(val_auc_hist)) + 1
    best_val_auc_p2 = float(np.max(val_auc_hist))
    final_val_auc_p2 = float(val_auc_hist[-1])
    # best_val_auc from Phase1 may or may not be in scope; guard it
    try:
        improvement = (best_val_auc_p2 - best_val_auc) * 100.0
    except Exception:
        improvement = None
    print(f"\n  Results:")
    print(f"    - Best epoch: {best_epoch_p2}")
    print(f"    - Best val_auc: {best_val_auc_p2:.4f}")
    print(f"    - Final val_auc: {final_val_auc_p2:.4f}")
    if improvement is not None:
        print(f"    - Improvement over Phase 1: {improvement:.2f}%")
else:
    print("\n  Results: no val_auc history found (check metric names).")

# Plot Phase 2 history
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(history_phase2.history.get('loss', []), label='Train Loss')
plt.plot(history_phase2.history.get('val_loss', []), label='Val Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Phase 2: Loss'); plt.legend(); plt.grid(True)

plt.subplot(1,3,2)
plt.plot(history_phase2.history.get('auc', []), label='Train AUC')
plt.plot(history_phase2.history.get('val_auc', []), label='Val AUC')
plt.xlabel('Epoch'); plt.ylabel('AUC'); plt.title('Phase 2: AUC'); plt.legend(); plt.grid(True)

plt.subplot(1,3,3)
plt.plot(history_phase2.history.get('binary_accuracy', []), label='Train Acc')
plt.plot(history_phase2.history.get('val_binary_accuracy', []), label='Val Acc')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.title('Phase 2: Accuracy'); plt.legend(); plt.grid(True)

plt.tight_layout()
plt.savefig('phase2_training_history.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n  ✓ Training plot saved: phase2_training_history.png")

# ---------------------------
# Load best model from Phase2 checkpoint (supply custom loss)
# ---------------------------
print("\n✓ Loading best model from Phase 2 checkpoint...")
try:
    model = tf.keras.models.load_model('phase2_best_model.h5', custom_objects={'loss': loss_fn})
    print("  ✓ Loaded best model from Phase 2 checkpoint")
except Exception as e:
    print("  ⚠️ Could not load model with custom loss. Continuing with in-memory model. Exception:", str(e))

print("\n✓ Phase 2 Complete!")
````

````python
# ================================================================================
# CELL 15: TEMPERATURE SCALING CALIBRATION (PER-CLASS, LOGIT-CORRECTED)
# ================================================================================
print("="*80)
print("CELL 15: TEMPERATURE SCALING CALIBRATION (PER-CLASS)")
print("="*80)

print("\n✓ Performing temperature scaling calibration...")

# Predict on validation set
val_preds = model.predict(val_dataset, verbose=1)
print(f"  - Validation predictions shape: {val_preds.shape}")
print(f"  - Validation labels shape: {val_labels.shape}")

# Convert probabilities → logits safely
def prob_to_logit(p):
    p = np.clip(p, 1e-7, 1 - 1e-7)
    return np.log(p / (1 - p))

val_logits = prob_to_logit(val_preds)

temperatures = np.ones(NUM_CLASSES)
losses = np.zeros(NUM_CLASSES)

# Temperature range
temp_range = np.arange(0.5, 5.0, 0.1)

print("\n✓ Calibrating each disease independently...")

for i in range(NUM_CLASSES):

    y_true = val_labels[:, i]
    y_logit = val_logits[:, i]

    best_temp = 1.0
    best_loss = float('inf')

    for T in temp_range:
        # Apply temperature scaling
        scaled_logits = y_logit / T
        scaled_probs = 1 / (1 + np.exp(-scaled_logits))

        # Clip for safety
        scaled_probs = np.clip(scaled_probs, 1e-7, 1 - 1e-7)

        # Binary crossentropy
        bce = -np.mean(
            y_true * np.log(scaled_probs) +
            (1 - y_true) * np.log(1 - scaled_probs)
        )

        if bce < best_loss:
            best_loss = bce
            best_temp = T

    temperatures[i] = best_temp
    losses[i] = best_loss
    print(f"  - {DISEASE_CLASSES[i]}: best T = {best_temp:.2f}, loss = {best_loss:.4f}")

print("\n==================== CALIBRATION SUMMARY ====================")
print(f"{'Disease':<22} {'Temp':>10} {'Loss':>10}")
print("-" * 50)
for i, d in enumerate(DISEASE_CLASSES):
    print(f"{d:<22} {temperatures[i]:>10.2f} {losses[i]:>10.4f}")

print("-" * 50)
print(f"{'MEAN':<22} {np.mean(temperatures):>10.2f} {np.mean(losses):>10.4f}")

# Save calibration parameters
calibration_dict = {
    'temperatures': temperatures.tolist(),
    'losses': losses.tolist()
}

with open("temperature_scaling.json", "w") as f:
    json.dump(calibration_dict, f, indent=4)

print("\n✓ Calibration parameters saved: temperature_scaling.json")
print("✓ Temperature Scaling Complete!")
````

````python
# ================================================================================
# CELL 16: FINAL TEST SET EVALUATION (WITH THRESHOLDS + TEMPERATURE CALIBRATION)
# ================================================================================
print("="*80)
print("CELL 16: FINAL TEST SET EVALUATION")
print("="*80)

print("\n✓ Generating predictions on test set...")
test_predictions = model.predict(test_dataset, verbose=1)
print(f"  - Test predictions shape: {test_predictions.shape}")

# ------------------------------------------------------------------------------
# STEP 1: Convert probabilities -> logits -> apply T per class -> convert back
# ------------------------------------------------------------------------------
def prob_to_logit(p):
    p = np.clip(p, 1e-7, 1 - 1e-7)
    return np.log(p / (1 - p))

def logit_to_prob(x):
    return 1 / (1 + np.exp(-x))

print("\n✓ Applying per-class temperature scaling...")

test_logits = prob_to_logit(test_predictions)

calibrated_probs = np.zeros_like(test_predictions)

for i in range(NUM_CLASSES):
    T = temperatures[i]      # from Cell 15
    scaled_logits = test_logits[:, i] / T
    calibrated_probs[:, i] = logit_to_prob(scaled_logits)

# ------------------------------------------------------------------------------
# STEP 2: Apply optimized thresholds (from Cell 14)
# ------------------------------------------------------------------------------
print("\n✓ Applying per-class thresholds...")

test_predictions_binary = np.zeros_like(calibrated_probs)

for i in range(NUM_CLASSES):
    th = optimal_thresholds[i]   # from Cell 14
    test_predictions_binary[:, i] = (calibrated_probs[:, i] >= th).astype(int)

# ------------------------------------------------------------------------------
# STEP 3: Compute metrics
# ------------------------------------------------------------------------------
print("\n✓ Computing metrics on test set...")

auc_scores = np.zeros(NUM_CLASSES)
f1_scores_per_disease = np.zeros(NUM_CLASSES)
recall_scores = np.zeros(NUM_CLASSES)
precision_scores = np.zeros(NUM_CLASSES)

for i in range(NUM_CLASSES):

    # AUC (only if both classes present)
    if len(np.unique(test_labels[:, i])) > 1:
        auc_scores[i] = roc_auc_score(test_labels[:, i], calibrated_probs[:, i])
    else:
        auc_scores[i] = np.nan

    # F1 / Recall / Precision using binary predictions
    f1_scores_per_disease[i] = f1_score(
        test_labels[:, i],
        test_predictions_binary[:, i],
        zero_division=0
    )
    recall_scores[i] = recall_score(
        test_labels[:, i],
        test_predictions_binary[:, i],
        zero_division=0
    )
    precision_scores[i] = precision_score(
        test_labels[:, i],
        test_predictions_binary[:, i],
        zero_division=0
    )

# ------------------------------------------------------------------------------
# STEP 4: Aggregate Metrics
# ------------------------------------------------------------------------------
mean_auc = np.nanmean(auc_scores)
mean_f1 = np.mean(f1_scores_per_disease)
mean_recall = np.mean(recall_scores)
mean_precision = np.mean(precision_scores)

# ------------------------------------------------------------------------------
# STEP 5: Display Results
# ------------------------------------------------------------------------------
print("\n" + "="*80)
print("FINAL TEST SET RESULTS")
print("="*80)

print(f"\n{'Disease':<20} {'AUC':>10} {'F1':>10} {'Recall':>10} {'Precision':>10}")
print("-" * 70)

for i, disease in enumerate(DISEASE_CLASSES):
    auc_str = f"{auc_scores[i]:.4f}" if not np.isnan(auc_scores[i]) else "NaN"
    print(f"{disease:<20} {auc_str:>10} {f1_scores_per_disease[i]:>10.4f} "
          f"{recall_scores[i]:>10.4f} {precision_scores[i]:>10.4f}")

print("-" * 70)
print(f"{'MEAN':<20} {mean_auc:>10.4f} {mean_f1:>10.4f} "
      f"{mean_recall:>10.4f} {mean_precision:>10.4f}")

print("\n✓ Evaluation complete!")

# ------------------------------------------------------------------------------
# STEP 6: Save CSV Results
# ------------------------------------------------------------------------------
results_df = pd.DataFrame({
    'Disease': DISEASE_CLASSES,
    'AUC': auc_scores,
    'F1_Score': f1_scores_per_disease,
    'Recall': recall_scores,
    'Precision': precision_scores,
    'Threshold': optimal_thresholds,
    'Temperature': temperatures
})

results_df.to_csv("test_results.csv", index=False)
print("\n✓ Results saved: test_results.csv")
````

````python
# ================================================================================
# CELL 19: CREATE INFERENCE FUNCTION (FINAL WORKING VERSION FOR NIH FOLDER SPLITS)
# ================================================================================
print("="*80)
print("CELL 19: CREATE INFERENCE FUNCTION (FINAL VERSION)")
print("="*80)

import cv2
import numpy as np
import os


# ------------------------------------------------------------------------------
# BUILD IMAGE FOLDER LIST
# ------------------------------------------------------------------------------
IMAGE_FOLDERS = [os.path.join(BASE_DIR, f"images_{i:03d}", "images")
                 for i in range(1, 13)]


# ------------------------------------------------------------------------------
# SUPPORT FUNCTIONS
# ------------------------------------------------------------------------------
def prob_to_logit(p):
    p = np.clip(p, 1e-7, 1 - 1e-7)
    return np.log(p / (1 - p))

def logit_to_prob(l):
    return 1 / (1 + np.exp(-l))


# ------------------------------------------------------------------------------
# RESOLVE IMAGE LOCATION
# ------------------------------------------------------------------------------
def resolve_image_path(filename):
    """Finds image in any of the NIH images_XXX/images folders."""
    
    # Case 1: Provided absolute path
    if os.path.exists(filename):
        return filename
    
    # Case 2: Try BASE_DIR/filename
    path2 = os.path.join(BASE_DIR, filename)
    if os.path.exists(path2):
        return path2
    
    # Case 3: Search inside all images_XXX/images folders
    for folder in IMAGE_FOLDERS:
        candidate = os.path.join(folder, filename)
        if os.path.exists(candidate):
            return candidate
    
    raise FileNotFoundError(f"Image not found in any dataset folder: {filename}")


# ------------------------------------------------------------------------------
# MAIN INFERENCE FUNCTION
# ------------------------------------------------------------------------------
def predict_on_image(
    image_filename,
    model,
    thresholds,
    temperatures,
    img_size=IMG_SIZE,
    disease_classes=DISEASE_CLASSES
):
    """Predict diseases for a single chest X-ray."""
    
    # Find image location
    img_path = resolve_image_path(image_filename)
    
    # Load and preprocess
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (img_size, img_size))
    img = img.astype(np.float32) / 255.0
    img = np.stack([img, img, img], axis=-1)
    img = np.expand_dims(img, axis=0)
    
    # Raw prediction
    pred = model.predict(img, verbose=0)[0]   # shape: (14,)
    pred = np.clip(pred, 1e-7, 1-1e-7)
    
    # Temperature scaling (per class)
    logits = prob_to_logit(pred)
    calibrated = np.zeros_like(pred)
    
    for i in range(len(pred)):
        scaled = logits[i] / temperatures[i]
        calibrated[i] = logit_to_prob(scaled)
    
    # Thresholding
    binary = (calibrated >= thresholds).astype(int)
    
    # Build result
    result = {
        "image": img_path,
        "predictions": {}
    }
    
    for i, disease in enumerate(disease_classes):
        result["predictions"][disease] = {
            "probability": float(calibrated[i]),
            "predicted": bool(binary[i]),
            "threshold": float(thresholds[i]),
            "temperature": float(temperatures[i])
        }
    
    return result


print("\n✓ Inference function created successfully!")
print("✓ All NIH image folders integrated!")
print("✓ FileNotFoundError problem solved!")

# ------------------------------------------------------------------------------
# TEST ON FIRST TEST IMAGE
# ------------------------------------------------------------------------------
example_img = test_images[0]
print(f"\nTesting inference on: {example_img}")

results = predict_on_image(example_img, model, optimal_thresholds, temperatures)

print("\nPositive predictions:")
found = False
for disease, info in results["predictions"].items():
    if info["predicted"]:
        print(f"  ✓ {disease:<20} {info['probability']:.4f} (thr={info['threshold']:.3f})")
        found = True
if not found:
    print("  (No findings detected)")

print("\n✓ Inference Test Complete!")
````

````python
pdf_generator_code = '''
from reportlab.lib.pagesizes import letter, A4
from reportlab.lib.units import inch
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph, Spacer, Image, PageBreak
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.enums import TA_CENTER, TA_LEFT, TA_RIGHT
from datetime import datetime
import io

class MedicalReportGenerator:
    """Generate professional medical PDF reports"""
    
    def __init__(self):
        self.styles = getSampleStyleSheet()
        self._create_custom_styles()
    
    def _create_custom_styles(self):
        """Create custom paragraph styles"""
        self.styles.add(ParagraphStyle(
            name='CustomTitle',
            parent=self.styles['Heading1'],
            fontSize=24,
            textColor=colors.HexColor('#1e3a8a'),
            spaceAfter=30,
            alignment=TA_CENTER,
            fontName='Helvetica-Bold'
        ))
        
        self.styles.add(ParagraphStyle(
            name='SectionHeader',
            parent=self.styles['Heading2'],
            fontSize=16,
            textColor=colors.HexColor('#2563eb'),
            spaceAfter=12,
            spaceBefore=12,
            fontName='Helvetica-Bold'
        ))
        
        self.styles.add(ParagraphStyle(
            name='BodyText',
            parent=self.styles['Normal'],
            fontSize=11,
            leading=14,
            spaceAfter=10
        ))
    
    def generate_report(self, results, image_path, patient_info, output_path):
        """Generate comprehensive medical report"""
        
        # Create PDF document
        doc = SimpleDocTemplate(output_path, pagesize=letter,
                                rightMargin=0.75*inch, leftMargin=0.75*inch,
                                topMargin=0.75*inch, bottomMargin=0.75*inch)
        
        story = []
        
        # Header
        story.append(Paragraph("CHEST X-RAY ANALYSIS REPORT", self.styles['CustomTitle']))
        story.append(Spacer(1, 0.2*inch))
        
        # Hospital/Clinic Info (Optional)
        header_data = [
            ['Report ID:', patient_info.get('report_id', 'N/A')],
            ['Date:', datetime.now().strftime('%B %d, %Y at %I:%M %p')],
            ['Patient ID:', patient_info.get('patient_id', 'N/A')],
            ['Patient Name:', patient_info.get('patient_name', 'Anonymous')],
            ['Age/Sex:', f"{patient_info.get('age', 'N/A')} / {patient_info.get('sex', 'N/A')}"]
        ]
        
        header_table = Table(header_data, colWidths=[1.5*inch, 4.5*inch])
        header_table.setStyle(TableStyle([
            ('ALIGN', (0, 0), (-1, -1), 'LEFT'),
            ('FONTNAME', (0, 0), (0, -1), 'Helvetica-Bold'),
            ('FONTSIZE', (0, 0), (-1, -1), 10),
            ('BOTTOMPADDING', (0, 0), (-1, -1), 8),
            ('GRID', (0, 0), (-1, -1), 0.5, colors.grey)
        ]))
        
        story.append(header_table)
        story.append(Spacer(1, 0.3*inch))
        
        # Analysis Results Section
        story.append(Paragraph("AI-ASSISTED DIAGNOSTIC FINDINGS", self.styles['SectionHeader']))
        story.append(Spacer(1, 0.1*inch))
        
        if len(results) == 0:
            story.append(Paragraph(
                "✓ <b>No abnormalities detected.</b> The AI analysis did not identify any significant pathological findings.",
                self.styles['BodyText']
            ))
        else:
            story.append(Paragraph(
                f"⚠ <b>Found {len(results)} potential abnormality(ies):</b>",
                self.styles['BodyText']
            ))
            story.append(Spacer(1, 0.1*inch))
            
            # Results table
            table_data = [['Disease', 'Confidence', 'Threshold', 'Status']]
            
            for res in results:
                confidence_pct = f"{res['confidence']*100:.1f}%"
                threshold_pct = f"{res['threshold']*100:.1f}%"
                status = "DETECTED" if res['detected'] else "Not Detected"
                
                table_data.append([
                    res['disease'],
                    confidence_pct,
                    threshold_pct,
                    status
                ])
            
            results_table = Table(table_data, colWidths=[2.5*inch, 1.3*inch, 1.3*inch, 1.5*inch])
            results_table.setStyle(TableStyle([
                ('BACKGROUND', (0, 0), (-1, 0), colors.HexColor('#2563eb')),
                ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),
                ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
                ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                ('FONTSIZE', (0, 0), (-1, 0), 11),
                ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
                ('BACKGROUND', (0, 1), (-1, -1), colors.beige),
                ('GRID', (0, 0), (-1, -1), 1, colors.black),
                ('FONTSIZE', (0, 1), (-1, -1), 10),
                ('ROWBACKGROUNDS', (0, 1), (-1, -1), [colors.white, colors.lightgrey])
            ]))
            
            story.append(results_table)
        
        story.append(Spacer(1, 0.3*inch))
        
        # X-Ray Image
        story.append(Paragraph("CHEST X-RAY IMAGE", self.styles['SectionHeader']))
        story.append(Spacer(1, 0.1*inch))
        
        img = Image(image_path, width=4*inch, height=4*inch)
        story.append(img)
        story.append(Spacer(1, 0.3*inch))
        
        # Clinical Notes
        story.append(Paragraph("CLINICAL INTERPRETATION", self.styles['SectionHeader']))
        story.append(Paragraph(
            "This report is generated using a DenseNet121-based deep learning model trained on the "
            "NIH Chest X-ray14 dataset. The model provides probability scores for 14 different thoracic diseases. "
            "<b>This is an AI-assisted screening tool and should be reviewed by a qualified radiologist "
            "before making any clinical decisions.</b>",
            self.styles['BodyText']
        ))
        
        story.append(Spacer(1, 0.2*inch))
        
        # Recommendations
        if len(results) > 0:
            story.append(Paragraph("RECOMMENDATIONS", self.styles['SectionHeader']))
            story.append(Paragraph(
                "• Follow-up with a board-certified radiologist for comprehensive evaluation<br/>"
                "• Clinical correlation with patient history and symptoms recommended<br/>"
                "• Consider additional imaging studies if clinically indicated<br/>"
                "• Correlate findings with physical examination and laboratory results",
                self.styles['BodyText']
            ))
        
        story.append(Spacer(1, 0.4*inch))
        
        # Footer
        story.append(Paragraph("_" * 80, self.styles['BodyText']))
        story.append(Paragraph(
            "<i>This report is computer-generated and does not replace professional medical advice. "
            "For questions or concerns, please consult your healthcare provider.</i>",
            ParagraphStyle('Footer', parent=self.styles['Normal'], fontSize=9, 
                           textColor=colors.grey, alignment=TA_CENTER)
        ))
        
        # Build PDF
        doc.build(story)
        
        return output_path
'''

# ✅ FIX: Specify encoding as 'utf-8'
with open('streamlit_app/pdf_generator.py', 'w', encoding='utf-8') as f:
    f.write(pdf_generator_code)

print("✅ pdf_generator.py created!")
````

````python
app_code = '''
import streamlit as st
import numpy as np
import cv2
from PIL import Image
import io
import os
from datetime import datetime
from model_loader import ChestXRayModel
from pdf_generator import MedicalReportGenerator

# Page configuration
st.set_page_config(
    page_title="Chest X-Ray Disease Classifier",
    page_icon="🩺",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS for dark theme matching your screenshots
st.markdown("""
    <style>
    /* Main background */
    .stApp {
        background-color: #0f172a;
    }
    
    /* Headers */
    h1, h2, h3 {
        color: #f1f5f9 !important;
    }
    
    /* File uploader */
    [data-testid="stFileUploader"] {
        background-color: #1e293b;
        border: 2px dashed #475569;
        border-radius: 10px;
        padding: 2rem;
    }
    
    /* Success/Warning boxes */
    .stAlert {
        background-color: #1e293b;
        border-radius: 8px;
    }
    
    /* Buttons */
    .stButton > button {
        background-color: #3b82f6;
        color: white;
        border-radius: 8px;
        padding: 0.5rem 2rem;
        font-weight: 600;
        border: none;
        transition: all 0.3s;
    }
    
    .stButton > button:hover {
        background-color: #2563eb;
        box-shadow: 0 4px 12px rgba(59, 130, 246, 0.4);
    }
    
    /* Disease cards */
    .disease-card {
        background: linear-gradient(135deg, #dc2626 0%, #991b1b 100%);
        padding: 1.5rem;
        border-radius: 12px;
        margin: 0.5rem 0;
        box-shadow: 0 4px 6px rgba(0, 0, 0, 0.3);
    }
    
    .disease-name {
        font-size: 1.2rem;
        font-weight: bold;
        color: white;
        margin-bottom: 0.5rem;
    }
    
    .confidence-grid {
        display: grid;
        grid-template-columns: 1fr 1fr;
        gap: 1rem;
        margin-top: 0.75rem;
    }
    
    .metric-box {
        background-color: rgba(0, 0, 0, 0.3);
        padding: 0.75rem;
        border-radius: 8px;
    }
    
    .metric-label {
        font-size: 0.85rem;
        color: #cbd5e1;
        margin-bottom: 0.25rem;
    }
    
    .metric-value {
        font-size: 1.8rem;
        font-weight: bold;
        color: white;
    }
    
    /* Success message */
    .success-box {
        background: linear-gradient(135deg, #059669 0%, #047857 100%);
        padding: 1.5rem;
        border-radius: 12px;
        color: white;
        font-size: 1.1rem;
        text-align: center;
        margin: 1rem 0;
    }
    </style>
""", unsafe_allow_html=True)

# Initialize session state
if 'model' not in st.session_state:
    st.session_state.model = None
if 'uploaded_image' not in st.session_state:
    st.session_state.uploaded_image = None
if 'results' not in st.session_state:
    st.session_state.results = None

@st.cache_resource
def load_model():
    """Load the trained model"""
    try:
        model = ChestXRayModel(
            model_path='../final_densenet121_model.h5',
            metadata_path='../model_metadata.json',
            thresholds_path='../optimal_thresholds.npy'
        )
        return model
    except Exception as e:
        st.error(f"Error loading model: {str(e)}")
        return None

def main():
    # Header with icon
    st.markdown("""
        <h1 style='text-align: center; margin-bottom: 0;'>
            🩺 Chest X-Ray Disease Classifier
        </h1>
    """, unsafe_allow_html=True)
    
    st.markdown("""
        <p style='text-align: center; color: #94a3b8; font-size: 1.1rem; margin-top: 0.5rem;'>
            Upload a chest X-ray image (PNG or JPG) to detect 14 different abnormalities.
        </p>
    """, unsafe_allow_html=True)
    
    # Load model
    if st.session_state.model is None:
        with st.spinner("🔄 Loading AI model..."):
            st.session_state.model = load_model()
    
    if st.session_state.model is None:
        st.error("❌ Failed to load model. Please check file paths.")
        return
    
    # Sidebar
    with st.sidebar:
        st.markdown("### ⚙️ Settings")
        
        # Patient information
        st.markdown("#### Patient Information (Optional)")
        patient_name = st.text_input("Patient Name", placeholder="Enter name")
        patient_id = st.text_input("Patient ID", placeholder="Enter ID")
        age = st.number_input("Age", min_value=0, max_value=120, value=0)
        sex = st.selectbox("Sex", ["Select", "Male", "Female", "Other"])
        
        st.markdown("---")
        
        # Model info
        st.markdown("#### 🤖 Model Information")
        st.info("""
        **Architecture:** DenseNet121  
        **Classes:** 14 thoracic diseases  
        **Dataset:** NIH Chest X-ray14  
        **Mean AUC:** 0.7738
        """)
        
        st.markdown("---")
        
        # About
        st.markdown("#### ℹ️ About")
        st.markdown("""
        This AI-powered tool assists in screening chest X-rays for:
        - Atelectasis
        - Cardiomegaly
        - Effusion
        - Infiltration
        - Mass
        - Nodule
        - Pneumonia
        - Pneumothorax
        - Consolidation
        - Edema
        - Emphysema
        - Fibrosis
        - Pleural Thickening
        - Hernia
        """)
    
    # Main content area
    col1, col2 = st.columns([1, 1])
    
    with col1:
        st.markdown("### Choose an X-ray image...")
        
        # File uploader
        uploaded_file = st.file_uploader(
            "Drag and drop file here",
            type=['png', 'jpg', 'jpeg'],
            label_visibility="collapsed",
            help="Limit 200MB per file • PNG, JPG, JPEG"
        )
        
        if uploaded_file is not None:
            # Display uploaded image
            image = Image.open(uploaded_file)
            st.image(image, caption="Uploaded X-ray", use_container_width=True)
            
            # Convert to numpy array
            img_array = np.array(image)
            if len(img_array.shape) == 2:
                img_array = cv2.cvtColor(img_array, cv2.COLOR_GRAY2RGB)
            elif img_array.shape[2] == 4:
                img_array = cv2.cvtColor(img_array, cv2.COLOR_RGBA2RGB)
            
            st.session_state.uploaded_image = img_array
            
            # Analyze button
            if st.button("🔬 Analyze X-Ray", use_container_width=True):
                with st.spinner("🧠 AI is analyzing..."):
                    results, predictions = st.session_state.model.predict(img_array)
                    st.session_state.results = results
    
    with col2:
        st.markdown("### Analysis Results")
        
        if st.session_state.results is not None:
            results = st.session_state.results
            
            if len(results) == 0:
                # No abnormalities detected
                st.markdown("""
                    <div class='success-box'>
                        ✅ No abnormalities detected.
                    </div>
                """, unsafe_allow_html=True)
            else:
                # Show detected abnormalities
                st.markdown(f"""
                    <div style='background-color: #854d0e; padding: 1rem; border-radius: 8px; 
                                color: white; text-align: center; margin-bottom: 1rem;'>
                        ⚠️ Found {len(results)} potential abnormality(ies):
                    </div>
                """, unsafe_allow_html=True)
                
                # Display each disease
                for result in results:
                    disease = result['disease']
                    confidence = result['confidence'] * 100
                    threshold = result['threshold'] * 100
                    
                    st.markdown(f"""
                        <div class='disease-card'>
                            <div class='disease-name'>{disease}</div>
                            <div class='confidence-grid'>
                                <div class='metric-box'>
                                    <div class='metric-label'>Confidence</div>
                                    <div class='metric-value'>{confidence:.1f}%</div>
                                </div>
                                <div class='metric-box'>
                                    <div class='metric-label'>Threshold</div>
                                    <div class='metric-value'>{threshold:.1f}%</div>
                                </div>
                            </div>
                        </div>
                    """, unsafe_allow_html=True)
            
            # Generate PDF Report button
            st.markdown("---")
            if st.button("📄 Generate PDF Report", use_container_width=True, type="primary"):
                with st.spinner("📝 Generating professional report..."):
                    # Save uploaded image temporarily
                    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                    temp_img_path = f"uploads/xray_{timestamp}.png"
                    cv2.imwrite(temp_img_path, cv2.cvtColor(st.session_state.uploaded_image, cv2.COLOR_RGB2BGR))
                    
                    # Patient info
                    patient_info = {
                        'report_id': f"CXR-{timestamp}",
                        'patient_name': patient_name if patient_name else "Anonymous",
                        'patient_id': patient_id if patient_id else "N/A",
                        'age': age if age > 0 else "N/A",
                        'sex': sex if sex != "Select" else "N/A"
                    }
                    
                    # Generate PDF
                    pdf_path = f"reports/chest_xray_report_{timestamp}.pdf"
                    generator = MedicalReportGenerator()
                    generator.generate_report(results, temp_img_path, patient_info, pdf_path)
                    
                    # Download button
                    with open(pdf_path, 'rb') as f:
                        pdf_data = f.read()
                    
                    st.success("✅ Report generated successfully!")
                    st.download_button(
                        label="⬇️ Download PDF Report",
                        data=pdf_data,
                        file_name=f"chest_xray_report_{timestamp}.pdf",
                        mime="application/pdf",
                        use_container_width=True
                    )
        else:
            st.info("👆 Upload an image and click 'Analyze X-Ray' to see results.")

if __name__ == "__main__":
    main()
'''

# 🟢 FIX: Specify UTF-8 encoding when writing (this prevents UnicodeEncodeError on Windows)
with open('streamlit_app/app.py', 'w', encoding='utf-8') as f:
    f.write(app_code)

print("✅ app.py created!")
print("\n📁 All files created successfully!")
````
